```python
# Find the total number of orders placed by each customer (customer_id and count).
Q7='''select c.customer_id, count(o.customer_id) as number_of_order
from Customers as c 
left join Orders as o 
on c.customer_id=o.customer_id
group by c.customer_id;'''
df7=pd.read_sql(Q7,connection)
df7
```

```python
# Retrieve a list of all orders, including customer names and order status.

Q8='''select CONCAT(first_name,' ', last_name) as name ,order_id, status 
from Customers as c 
join Orders as o 
on c.customer_id=o.customer_id;'''
df8=pd.read_sql(Q8,connection)
df8
```

```python
# Find the total revenue (sum of all orders) generated.
Q9='''select round(sum(total_amount),2) as total_revenue
from Orders;'''
df9=pd.read_sql(Q9,connection)
df9
```

```python
Q10='''select c.customer_id, count(o.customer_id) as number_of_order
from Customers as c 
left join Orders as o 
on c.customer_id=o.customer_id
group by c.customer_id
having count(o.customer_id)>3;'''
df10=pd.read_sql(Q10,connection)
df10
```

```python
# Retrieve all completed payments and include the associated order details.

Q11='''select p.payment_id, o.order_date, o.total_amount, p.payment_method, p.payment_status
from Orders as o 
join Payment as p 
on o.order_id=p.order_id
where p.payment_status = 'Completed';'''
df11=pd.read_sql(Q11,connection)
df11
```

```python
# Find the top 5 customers who have spent the most on orders.

Q12='''select concat(first_name, ' ', last_name) as name, round(sum(total_amount),2) as total_amount 
from Customers as c 
join Orders as o 
on c.customer_id=o.customer_id
group by name 
order by total_amount desc 
limit 5;'''
df12=pd.read_sql(Q12,connection)
df12
```

```python
# List all orders where the payment was either ‘Pending’ or ‘Failed’.
Q13='''select o.order_id, payment_status 
from Orders as o 
join Payment as p 
on o.order_id= p.order_id 
where payment_status in ('Failed', 'Pending')
order by o.order_id;'''
df13=pd.read_sql(Q13,connection)
df13
```

```python
# Find the most popular product category based on the number of orders.
Q14='''select category , count(order_id) as number_of_order
from Products as p 
join Orders as o 
on p.product_id=o.product_id 
group by category
order by number_of_order desc
limit 1;'''
df14=pd.read_sql(Q14,connection)
df14
```

```python
# Identify orders that were placed but have no successful payment recorded.

Q15='''select o.order_id, payment_status 
from Orders as o 
join Payment as p 
on o.order_id=p.order_id
where payment_status!='Completed'
order by o.order_id;'''
df15=pd.read_sql(Q15,connection)
df15
```

```python
# Calculate the average order value (AOV) per customer.
Q16='''select c.customer_id, round(avg(total_amount),2) as avg_amount
from Customers as c 
join Orders as o 
on c.customer_id=o.customer_id 
group by c.customer_id;'''
df16=pd.read_sql(Q16,connection)
df16
```

```python
# Identify which payment method is most commonly used.

Q17='''select payment_method, count(payment_method) as count
from Payment
group by payment_method
order by count desc
limit 1;'''
df17=pd.read_sql(Q17,connection)
df17
```

```python
# Determine the percentage of orders that were canceled.
Q18='''SELECT 
    (COUNT(CASE WHEN status = 'Cancelled' THEN 1 END) * 100.0 / COUNT(*)) AS canceled_percentage
FROM Orders;'''
df18=pd.read_sql(Q18,connection)
df18
```

```python
# Find out the month in which the most orders were placed.
Q19='''select MONTH(order_date) as Month , count(order_id) as count
from Orders 
group by MONTH(order_date)
order by MONTH(order_date), count desc
limit 1;'''
df19=pd.read_sql(Q19,connection)
df19
```
